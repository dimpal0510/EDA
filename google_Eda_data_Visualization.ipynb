**Apps dataset columns descriptions:**

This file contains all the details of the apps on Google Play. There are 9 features that describe a given app.
<ul>
    <li><b>App:</b> Name of the app</li>
    <li><b>Category:</b> Category of the app. Some examples are: ART_AND_DESIGN, FINANCE, COMICS, BEAUTY etc.</li>
    <li><b>Rating:</b> The current average rating (out of 5) of the app on Google Play</li>
    <li><b>Reviews:</b> Number of user reviews given on the app</li>
    <li><b>Size:</b> Size of the app in MB (megabytes)</li>
    <li><b>Installs:</b> Number of times the app was downloaded from Google Play</li>
    <li><b>Type:</b> Whether the app is paid or free</li>
    <li><b>Price:</b> Price of the app in US$</li>
    <li><b>Last Updated:</b> Date on which the app was last updated on Google Play</li>
</ul>

To  Visualize missing number data, let me use seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#Let's set the theme of plots.
sns.set_theme()
sns.set(rc = {"figure.dpi":300, "figure.figsize":(12,9)})
```

To see missing data, I'm going to use the heatmap method.

```python
sns.heatmap(df.isnull(),cbar = False)
```

```python
rating_median = df["Rating"].median()
print(rating_median)
df["Rating"].fillna(rating_median, inplace = True)
```

Let's remove the other missing data as it is very little.

```python
df.dropna(inplace = True)
```

let's take alook on missing data again

```python
df.isnull().sum().sum()
```

```python
df.info()
```

let's convert the type of reviews into numeric.

```python
df["Reviews"] = df["Reviews"].astype("int64")
```

```python
df["Reviews"].describe().round()
```

Let's remove the M and k in the size of column.

```python
df["Size"].replace("M","", regex = True, inplace = True)
df["Size"].replace("k","", regex = True, inplace = True)
```

```python
df["Size"].unique()
```

```python
size_median = df[df["Size"] != "Varies with device"]["Size"].astype(float).median()
size_median
```

```python
df["Size"].replace("Varies with device", size_median, inplace = True)
```

```python
df.Size = pd.to_numeric(df.Size)
```

```python
df.Size.head()
```

```python
df.Size.describe().round()
```

# Handling the installs column

```python
df["Installs"].unique()
```

```python
df.Installs = df.Installs.apply(lambda x:x.replace("+",""))
df.Installs = df.Installs.apply(lambda x:x.replace(",",""))
df.Installs = df.Installs.apply(lambda x: int(x))
```

```python
df["Installs"].unique()
```

# Handaling The price column

```python
df["Price"].unique()
```

Let's remove the $ symbol in this column

```python
df.Price = df.Price.apply(lambda x:x.replace("$",""))
df.Price = df.Price.apply(lambda x:float(x))
```

```python
df["Price"].unique()
```

# Handling the genres column

```python
len(df["Genres"].unique())
```

```python
df["Genres"].head(10)
```

The data is divided between genres and subgenres, let's only set ny genre.

```python
df["Genres"] = df["Genres"].str.split(";").str[0]
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].unique()
```

```python
df["Genres"].value_counts()
```

Let's combine the music and audio with the music

```python
df["Genres"].replace("Music & Audio", "Music", inplace = True)
```

# Handling the last updated column

```python
df["Last_Updated"].head()
```

```python
df["Last_Updated"] = pd.to_datetime(df["Last_Updated"])
```

# Bar Plot for the category column with the number of the categoris

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("googleplaystore.csv")


cat_num = df["Category"].value_counts()

cat_df = cat_num.reset_index()
cat_df.columns = ['Category', 'Count'] # Rename columns for clarity

sns.barplot(x='Count', y='Category', data=cat_df)
plt.title("The number of categories", size=20)
```
